```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
# Задаем рандомное распределение точек
xi = np.random.uniform(-8, 8, 200)
yi = 1/(1+np.exp(-np.random.uniform(0.5, 1.5, 200)*xi-np.random.uniform(-1, 1, 200)))
plt.plot(xi,yi, marker="o", ls="")

# Сигмоидальная функция
def myFunc(x, a, b):
    return (np.exp(-(a*x+b))+ 1)**(-1)

# Узнаем параметры a, b
popt, pcov = curve_fit(myFunc, xi, yi)

# Строим сигмоидальную функцию
x = np.linspace(-8, 8, 200)
y=myFunc(x, a = popt[0], b = popt[1])

plt.plot(x, y, c='r')
plt.show()
```
